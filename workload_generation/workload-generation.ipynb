```python
import subprocess
import time
```

```python
#automatically change the workload
def auto_stress_function(runtime, time_duration):
    endpoint = '145.100.135.52'
    timestep_duration = 0.1
    users = [10, 20, 30, 40, 50, 40, 30, 20, 10]
    rates = [1, 3, 5, 7, 9, 7, 5, 3, 1]
    idx = 0
    while True:
        idx %= 9
        print('start: locust -f ./locust.py --headless -u {} -r {} -H $1 -t {}'.format(users[idx], rates[idx], runtime))
        locust_process = subprocess.Popen('locust-stress.sh {} {} {} {}'
                   .format(endpoint, users[idx], rates[idx], runtime), shell=True, close_fds=True)
        time.sleep(time_duration)
        locust_process.terminate()
        locust_process.kill()
        idx += 1

    return 0
```

```python
#just one fixed requesting workload
def stress_function(user, rate, runtime, time_duration):
    endpoint = '145.100.135.52'
    
    print('start: locust -f ./locust.py --headless -u {} -r {} -H {} -t {}'.format(user, rate, endpoint,
                                                                                   runtime))
    locust_process = subprocess.Popen('./locust-stress.sh {} {} {} {}'
               .format(endpoint, user, rate, runtime), shell=True, close_fds=True)
    time.sleep(time_duration)
    locust_process.terminate()
    locust_process.kill()
    print('Killed.')

    return 0
```
