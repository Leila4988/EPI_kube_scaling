```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.ticker as mticker  
```

## Obtain history data

```python
df = pd.read_csv('k8s_historical_discrete_states_five_actions.csv')
results = df[['timestep', 'reward', 'number_of_pods', 'latency', 'info', 'hpa_threshold']]
```

## Plot the tendency in training period with -u=40, action_space = 5

```python
podsTrainVis = results[91:][['timestep', 'number_of_pods', 'info']]
latencyTrainVis = results[91:][['timestep', 'latency', 'info']]
thesTrainVis = results[91:][['timestep', 'hpa_threshold', 'info']]
rewardTrainVis = results[91:][['timestep', 'reward', 'info']]
```

```python
plt.figure(dpi=120)
plt.xlabel('timestep')
plt.ylabel('number_of_pods')
plt.title('tendency of pods in training')
plt.plot(podsTrainVis['timestep'], podsTrainVis['number_of_pods'])
plt.show()
```

```python
plt.figure(dpi=120)
plt.xlabel('timestep')
plt.ylabel('hpa_threshold')
plt.title('tendency of hpa_threshold in training')
plt.plot(thesTrainVis['timestep'], thesTrainVis['hpa_threshold'])
plt.show()
```

```python
plt.figure(dpi=120)
plt.xlabel('timestep')
plt.ylabel('reward')
plt.title('tendency of reward in training')
plt.plot(rewardTrainVis['timestep'], rewardTrainVis['reward'])
plt.show()
```

```python
plt.figure(dpi=120)
plt.xlabel('timestep')
plt.ylabel('latency')
plt.title('tendency of latency in training')
plt.plot(latencyTrainVis['timestep'], latencyTrainVis['latency'])
```

## Plot the tendency in testing period with bursts

```python
podsTestVis = results[499:][['timestep', 'number_of_pods', 'info']]
latencyTestVis = results[499:][['timestep', 'latency', 'info']]
rewardTestVis = results[499:][['timestep', 'reward', 'info']]
thesTestVis = results[499:][['timestep', 'hpa_threshold', 'info']]
# podsTestVis = results[499:][['timestep', 'number_of_pods', 'info']]
# latencyTestVis = results[499:][['timestep', 'latency', 'info']]
# rewardTestVis = results[499:][['timestep', 'reward', 'info']]
# thesTestVis = results[499:][['timestep', 'hpa_threshold', 'info']]
```

```python
plt.figure(dpi=120)
plt.xlabel('timestep')
plt.ylabel('number_of_pods')
plt.title('tendency of pods in testing')
plt.plot(podsTestVis['timestep'], podsTestVis['number_of_pods'])
plt.show()
```

```python
plt.figure(dpi=120)
plt.xlabel('timestep')
plt.ylabel('reward')
plt.title('tendency of reward in testing')
plt.plot(rewardTestVis['timestep'], rewardTestVis['reward'])
```

```python
plt.figure(dpi=120)
plt.xlabel('timestep')
plt.ylabel('hpa_threshold')
plt.title('tendency of hpa_threshold in testing')
plt.plot(thesTestVis['timestep'], thesTestVis['hpa_threshold'])
plt.show()
```

```python
plt.figure(dpi=120)
plt.xlabel('timestep')
plt.ylabel('latency')
plt.title('tendency of latency in testing')
plt.plot(latencyTestVis['timestep'], latencyTestVis['latency'])
```

```python
plt.figure(dpi=120)
plt.xlabel('round')
plt.ylabel('hpa_threshold')
plt.title('tendency of hpa_threshold in testing')
plt.plot(thesTestVis['timestep'], thesTestVis['hpa_threshold']/20)
plt.plot(rewardTestVis['timestep'], rewardTestVis['reward']/10)
plt.plot(podsTestVis['timestep'], podsTestVis['number_of_pods'])
plt.plot(latencyTestVis['timestep'], latencyTestVis['latency'])
plt.legend(['hpa_threshold', 'reward', 'number_of_pods', 'latency'], loc=(1.04,0)) 
plt.show()
```

## Plot training in one figure

```python
plt.figure(figsize=(15, 12))
plt.subplot(2, 2, 1)
# plt.figure(dpi=120)
plt.xlabel('round')
plt.ylabel('number_of_pods')
plt.title('changes in number of pods during training')
plt.plot(podsTrainVis['timestep'], podsTrainVis['number_of_pods'])
plt.vlines(100, 0, max(podsTrainVis['number_of_pods']), colors = "r", linestyles = "dashed")
plt.subplot(2, 2, 2)
# plt.figure(dpi=120)
plt.xlabel('round')
plt.ylabel('hpa_threshold')
plt.title('changes in hpa_threshold during training')
# 添加此段代码即可
def to_percent(temp, position):
    return '%1.0f'%(temp) + '%'
plt.gca().yaxis.set_major_formatter(mticker.FuncFormatter(to_percent))
plt.plot(thesTrainVis['timestep'], thesTrainVis['hpa_threshold'])
plt.vlines(100, 0, max(thesTrainVis['hpa_threshold']), colors = "r", linestyles = "dashed")
# plt.show()
# plt.figure(figsize=(15, 5))
plt.subplot(2, 2, 3)
plt.xlabel('round')
plt.ylabel('reward')
plt.title('changes in rewards during training')
plt.plot(rewardTrainVis['timestep'], rewardTrainVis['reward'])
plt.vlines(100, 0, max(rewardTrainVis['reward']), colors = "r", linestyles = "dashed")
plt.subplot(2, 2, 4)
plt.xlabel('round')
plt.ylabel('latency/seconds')
plt.title('changes in latency during training')
plt.gca().yaxis.set_major_formatter(mticker.FormatStrFormatter('%.1f s'))
plt.plot(latencyTrainVis['timestep'], latencyTrainVis['latency'])
plt.vlines(100, 0, max(latencyTrainVis['latency']), colors = "r", linestyles = "dashed")
```
