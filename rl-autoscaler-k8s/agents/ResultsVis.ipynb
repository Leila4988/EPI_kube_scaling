```python
import pandas as pd
import matplotlib.pyplot as plt
plt.figure(figsize=(6, 8))
```

## Obtain history data

```python
df = pd.read_csv('k8s_historical_states_discrete.csv')
```

```python
results = df[['timestep', 'reward', 'number_of_pods', 'latency', 'info', 'hpa_threshold']]
```

## Plot the tendency in training period with -u=10

## Plot the tendency in training period with -u=40

```python
podsTrainVis = results[359:497][['timestep', 'number_of_pods', 'info']]
latencyTrainVis = results[359:497][['timestep', 'latency', 'info']]
thesTrainVis = results[359:497][['timestep', 'hpa_threshold', 'info']]
rewardTrainVis = results[359:497][['timestep', 'reward', 'info']]
```

```python
plt.figure(dpi=120)
plt.xlabel('timestep')
plt.ylabel('number_of_pods')
plt.title('tendency of pods in training')
plt.plot(podsTrainVis['timestep'], podsTrainVis['number_of_pods'])
plt.show()
```

```python
plt.figure(dpi=120)
plt.xlabel('timestep')
plt.ylabel('hpa_threshold')
plt.title('tendency of hpa_threshold in training')
plt.plot(thesTrainVis['timestep'], thesTrainVis['hpa_threshold'])
plt.show()
```

```python
plt.figure(dpi=120)
plt.xlabel('timestep')
plt.ylabel('reward')
plt.title('tendency of reward in training')
plt.plot(rewardTrainVis['timestep'], rewardTrainVis['reward'])
plt.show()
```

```python
plt.figure(dpi=120)
plt.xlabel('timestep')
plt.ylabel('latency')
plt.title('tendency of latency in training')
plt.plot(latencyTrainVis['timestep'], latencyTrainVis['latency'])
```

## Plot the tendency in testing period with bursts

```python
podsTestVis = results[499:][['timestep', 'number_of_pods', 'info']]
latencyTestVis = results[499:][['timestep', 'latency', 'info']]
rewardTestVis = results[499:][['timestep', 'reward', 'info']]
thesTestVis = results[499:][['timestep', 'hpa_threshold', 'info']]
```

```python
plt.figure(dpi=120)
plt.xlabel('timestep')
plt.ylabel('number_of_pods')
plt.title('tendency of pods in testing')
plt.plot(podsTestVis['timestep'], podsTestVis['number_of_pods'])
plt.show()
```

```python
plt.figure(dpi=120)
plt.xlabel('timestep')
plt.ylabel('reward')
plt.title('tendency of reward in testing')
plt.plot(rewardTestVis['timestep'], rewardTestVis['reward'])
```

```python
plt.figure(dpi=120)
plt.xlabel('timestep')
plt.ylabel('hpa_threshold')
plt.title('tendency of hpa_threshold in testing')
plt.plot(thesTestVis['timestep'], thesTestVis['hpa_threshold'])
plt.show()
```

```python
plt.figure(dpi=120)
plt.xlabel('timestep')
plt.ylabel('latency')
plt.title('tendency of latency in testing')
plt.plot(latencyTestVis['timestep'], latencyTestVis['latency'])
```

```python
plt.figure(dpi=200)
plt.xlabel('timestep')
plt.ylabel('hpa_threshold')
plt.title('tendency of hpa_threshold in testing')
plt.plot(thesTestVis['timestep'], thesTestVis['hpa_threshold']/20)
plt.plot(rewardTestVis['timestep'], rewardTestVis['reward']/10)
plt.plot(podsTestVis['timestep'], podsTestVis['number_of_pods'])
plt.plot(latencyTestVis['timestep'], latencyTestVis['latency'])
plt.legend(['hpa_threshold', 'reward', 'number_of_pods', 'latency']) 
plt.show()
```
